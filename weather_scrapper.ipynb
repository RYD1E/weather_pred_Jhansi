```python
import requests as rts
import json
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
default_values = {
    "hl":"uk",
    "hls":"uk",
    "hlsh":"uk",
}
for j in [2019,2020,2021,2022,2023]:
    for i in range(1,13):
        page=rts.get("https://www.timeanddate.com/weather/india/jhansi/historic?month="+str(i)+"&year="+str(j))
        soup=bs(page.content,"html.parser")
        df=pd.DataFrame(json.loads(soup.find_all("script",type="text/javascript")[1].text.split('detail":')[-1].split("]")[0]+"]",object_hook=lambda d: {**default_values, **d}))
        df['Month']=i
        df["Year"]=j
        df.to_csv(r"C:\Users\ayush\Desktop\study\weather"+str(i)+", "+str(j)+".csv")
        if j==2023 and i > 2:
            break
```

```python
df1=pd.DataFrame()
index=0
for j in [2019,2020,2021,2022,2023]:
    for i in range(1,13):
        if j==2023 and i > 2:
            break
        else:
            df=pd.read_csv(r"C:\Users\ayush\Desktop\study\Weather_data\weather"+str(i)+", "+str(j)+".csv")
            lis=[]
            for ii in df.ds:
                lis.append(ii.split(" ")[1])
            df["ds"]=lis
            df=df.drop(columns=["hl","hls","hlsh","desc","icon","date"],axis=1)
            df=df.fillna(1000000)
            for date in df.ds.unique():
                df1.loc[index,"day"]=int(date)
                df1.loc[index,"month"]=int(df.Month[0])
                df1.loc[index,"year"]=int(df.Year[0])
                df1.loc[index,"temp_max"]=int(df[df["ds"]== date].temp.max())
                df1.loc[index,"temp_min"]=int(df[df["ds"]== date].templow.min())
                df1.loc[index,"atm_avg"]=int(df[df["ds"]== date].baro.mean())
                df1.loc[index,"hum_max"]=int(df[df["ds"]== date].hum.max())
                df1.loc[index,"hum_min"]=int(df[df["ds"]== date].hum.min())
                index=index+1
df1.to_csv(r"C:\Users\ayush\Desktop\study\Weather_data\weather_finaldata.csv")
```
